### Exercise 4: Pie Chart with Percentages

```python
import matplotlib.pyplot as plt

# Step 1: Define labels and sizes
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Step 2: Create a pie chart with percentages
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'])

# Step 3: Add a title
plt.title('Pie Chart with Percentages')

# Show the plot
plt.show()
```
