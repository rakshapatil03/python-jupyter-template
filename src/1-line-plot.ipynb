### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt

# Step 1: Define x and y values
x = list(range(11))  # x values from 0 to 10
y = [i**2 for i in x]  # y values as the square of x

# Step 2: Create a line plot
plt.plot(x, y, marker='o', linestyle='-', color='blue', label='y = x²')

# Step 3: Customize the plot
plt.title('Line Plot Customization')
plt.xlabel('X Values')
plt.ylabel('Y Values (Squared)')
plt.grid(True)  # Add grid
plt.legend()  # Add legend

# Show the plot
plt.show()
```
