### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt

# Create a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# --- 1st Subplot: Line Plot ---
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]
axes[0, 0].plot(x, y, marker='o', linestyle='-', color='blue')
axes[0, 0].set_title('Line Plot')

# --- 2nd Subplot: Bar Plot ---
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
axes[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axes[0, 1].set_title('Bar Plot')

# --- 3rd Subplot: Scatter Plot ---
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axes[1, 0].scatter(x_scatter, y_scatter, color='red', marker='o')
axes[1, 0].set_title('Scatter Plot')

# --- 4th Subplot: Pie Chart ---
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'])
axes[1, 1].set_title('Pie Chart')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
